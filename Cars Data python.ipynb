# 3D Visualizatio  

```python
fig = plt.figure(figsize=(16,8))
ax= plt.axes(projection='3d')
ax.set_xlabel('EngineSize')
ax.set_ylabel('Horsepower')
ax.set_zlabel('Weight')

ax.scatter3D(car['EngineSize'],car['Horsepower'],car['Weight'],c='red')
```

```python
fig = plt.figure(figsize=(16,8))
ax= plt.axes(projection='3d')
ax.set_xlabel('Horsepower')
ax.set_ylabel('EngineSize')
ax.set_zlabel('Weight')

ax.scatter3D(car['Horsepower'],car['EngineSize'],car["Weight"],c='magenta')
```

```python
fig = plt.figure(figsize=(16,8))
ax= plt.axes(projection='3d')
ax.set_xlabel('Weight')
ax.set_ylabel('EngineSize')
ax.set_zlabel('Wheelbase')

ax.scatter3D(car['Weight'],car['EngineSize'],car['Wheelbase'],c='cyan')
```
